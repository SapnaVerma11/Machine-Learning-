```python

df.drop(columns = ['Id','PoolQC','Fence','MiscFeature','FireplaceQu','Alley','LotFrontage','BsmtFinType2'],inplace  = True)
df.shape
```

```python
plt.figure(figsize = (10,10))

sns.heatmap(df.isnull(),yticklabels=False)
```

```python
# filling numerical columns missing values with mean 
df.fillna(df.mean(),inplace = True)
df.isnull().sum()
```

```python
# filling categorucal columns missing values with mean 

df.fillna(method = 'bfill')
df.isnull().sum()
```

```python
plt.figure(figsize = (20,20))

sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df.shape
```

```python
# printing all categorical columns
obj_list = df.select_dtypes(include = "object").columns
obj_list
```

```python
# converting all categorical columns into numerical data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for feature in obj_list:
    df[feature] = le.fit_transform(df[feature].astype(str))
```

```python
df.head(10)
```

```python
df.dtypes
```

```python
# separting x and y variables
y = df.iloc[::,-1].values
y.shape
```

```python
x = df.iloc[:,:-1]
x.shape
```

```python
y = y.reshape(-1,1)
```

```python
# splitting data into training and testing 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#creating linear regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#fitting the training data
model.fit(x_train,y_train)
```

```python
#Predictions
ypredict = model.predict(x_test)
```

```python
#checking the r2score
from sklearn.metrics import r2_score
r2_score(y_test,ypredict)
```
