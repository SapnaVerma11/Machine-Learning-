```python
# splitting x and y variable
x = df.iloc[:,1:]
x.head()
```

```python
y = df.iloc[:,0]
y.head()
```

#### Feature Scaling

```python
from sklearn import preprocessing
x = preprocessing.StandardScaler().fit_transform(x)
x[0:2]
```

#### GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
k_range = list(range(1,31))
weight_options = ["uniform","distance"]
metric_options = ["euclidean","manhattan"]
param_grid = dict(n_neighbors = k_range,weights = weight_options,metric = metric_options)
```

```python
knn = KNeighborsClassifier()
```

```python
grid = GridSearchCV(knn,param_grid,cv = 5,verbose = 1,n_jobs = -1)
grid.fit(x,y)
```

```python
print(grid.best_estimator_)
print(grid.best_score_)
print(grid.best_params_)
```

```python
# train test splitting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# traning model
knn = KNeighborsClassifier(n_neighbors=11)
knn.fit(x_train,y_train)
ypredict = knn.predict(x_test)
```

#### checking accuracy 

```python
from sklearn.metrics import accuracy_score
acc  = accuracy_score(y_test,ypredict)
acc
```

#### performing confusion matrix


```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,ypredict)
cm
```
